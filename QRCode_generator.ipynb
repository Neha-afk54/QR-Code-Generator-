```python
import qrcode
import matplotlib.pyplot as plt

# Take user input
data = input("Enter the text or URL to generate QR Code: ")

# Generate QR code
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4
)

qr.add_data(data)
qr.make(fit=True)

# Create image
img = qr.make_image(fill_color="black", back_color="pink")

# Display QR code
plt.imshow(img)
plt.axis("off")
plt.title("Generated QR Code")
plt.show()

# Save QR code
img.save("user_qrcode.png")

print("QR Code generated and saved as 'user_qrcode.png'")
```
